```python
import glob
import pandas as pd
from datetime import datetime
```

```python
# Extract
def extract():
    extracted_data = pd.DataFrame(columns = ['car_model', 'year_of_manufacture', 'price', 'fuel'])
    # CSV
    for file in glob.glob("datasource/*.csv"):
        csv_file = pd.read_csv(file)
        extracted_data =pd.concat([extracted_data,csv_file])

    # json
    for file in glob.glob("datasource/*.json"):
        jason_file = pd.read_json(file,lines=True)
        extracted_data =pd.concat([extracted_data,jason_file])

    # xml

    return extracted_data
```
